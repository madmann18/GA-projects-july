# EDA Reflection

Moving forward,I am going to look at regression using both the white wine set and the wines set. I want to see if the mixed wine performs better or worse that just the white wine. I am using the white wine rather than red because there are more observations for this group.

I would like to use the type of wine as the variable I would like to classify when i do classification to see if the dataset can provide me with a good indicator of wine type.

```python
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
```

```python
#all wine dataset
#dropping variables to account for multicollinearity & quality (y variable)
winesreg = wines.drop([ 'quality','pH', 'free sulfur dioxide', 'volatile acidity'], axis = 1) 
#adding dummy variable for type
winesreg = pd.get_dummies(winesreg, drop_first = True)
X = winesreg
y = wines['quality']
X_train, X_test, y_train, y_test = train_test_split(X,y)
lm = LinearRegression()
lm.fit(X_train, y_train)
pred = lm.predict(X_test)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
print('mse: ', mse)
print ('rmse: ', rmse)
```

```python
print (lm.coef_)
print (lm.intercept_)
print (lm.score(X_test, y_test))
```

```python
#white wine dataset
#again dropping variables with correation & quality (y variable)
wwinesreg = wwines.drop(['quality','density', 'free sulfur dioxide','pH', 'alcohol'], axis = 1) 
#adding dummy variable for type
wwinesreg = pd.get_dummies(wwinesreg, drop_first = True)
X = wwinesreg
y = wwines['quality']
X_train, X_test, y_train, y_test = train_test_split(X,y)
lm = LinearRegression()
lm.fit(X_train, y_train)
pred = lm.predict(X_test)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)
print('mse: ', mse)
print ('rmse: ', rmse)
```

```python
print (lm.coef_)
print (lm.intercept_)
print (lm.score(X_test, y_test))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
winesclass = wines.drop([ 'type','pH', 'free sulfur dioxide', 'volatile acidity'], axis = 1) 
```

```python
X = winesclass
y = wines['type']
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
pipe = make_pipeline(PolynomialFeatures(), LogisticRegression())
params = {'polynomialfeatures__degree': [i for i in range(1,5)], 'logisticregression__C': [0.1, 1, 10, 50]}
grid = GridSearchCV(pipe, param_grid=params, cv = 3)
grid.fit(X_train, y_train)
best = grid.best_estimator_
pred = best.predict(X_test)
print (best.score(X_test, y_test))
```

```python
print(best)
```

```python
print(classification_report(pred, y_test))
```
